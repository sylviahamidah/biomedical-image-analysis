```python
# Read multi-dimensional data directly
import imageio.v2 as imageio

folder_path = 'E:\ITS_Stuff\Semester 6\pcm_v2\Week_9\dataset1_Thorax_Abdomen'
vol1 = imageio.volread(folder_path, format='DICOM')
n0, n1, n2 = vol1.shape
n0, n1, n2
```

```python
import os
os.listdir('E:\ITS_Stuff\Semester 6\pcm_v2\Week_9\dataset1_Thorax_Abdomen')
```

```python
# Sampling rate (in mm): physical space covered by each element.

d0, d1, d2 = vol1.meta['sampling']    
d0, d1, d2  #d0 = jarak antar slicing, d1 dan d2 ukuran 1 pixel
```

```python
# Field of view (in mm):  physical space covered along each axis.
n0 * d0, n1 * d1, n2 * d2
```

```python
import imageio
#vol3 = imageio.volread('D:\Teaching\Pengolahan Citra Medika\Database\dataset1_Thorax_Abdomen')
fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(20,10))
view_axial = vol1[20, :, :]
view_coronal = vol1[ :, 450, :]
view_sagittal = vol1[ :, :, 300]

axes[0].imshow(view_axial)
axes[1].imshow(view_coronal)
axes[2].imshow(view_sagittal)

for ax in axes:
    ax.axis('off')

plt.show()
```

```python
# Modifying the aspect ratio
view_sagittal = vol1[ :, :, 300]
d0, d1, d2 = vol1.meta['sampling']
d0, d1, d2
```

```python
asp1 = d0/d1
asp1
```

```python
plt.imshow(view_sagittal, aspect = asp1)
plt.axis('off')
plt.show()
```

```python
view_coronal = vol1[ :, 450, :]
asp2 = d0/d2
plt.imshow(view_coronal, aspect = asp2)
plt.axis('off')
plt.show()
```
